```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df
```

```python
df = df.drop_duplicates(subset="id")
```

```python
df
```

```python
price_cols = ["price", "cleaning_fee", "extra_people"]

for col in price_cols:
    if col in df.columns:
        df[col] = df[col].replace('[\$,]', '', regex=True)
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.head(2)
```

```python
df.columns
```

```python
list(df.columns)
```

```python
percent_cols = ["host_response_rate", "host_acceptance_rate"]

for col in percent_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)                  # ensure everything is string
            .str.replace("%", "", regex=False)
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df
```

```python
date_cols = ["last_review", "first_review", "host_since"]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
df.head(2)
```

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.head(2)
```

```python
if "reviews_per_month" in df.columns:
    df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
binary_map = {
    "t": 1, "f": 0,
    "True": 1, "False": 0
}

binary_cols = ["host_is_superhost", "instant_bookable"]

for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].map(binary_map)
```

```python
df.head(2)
```

```python
text_cols = ["room_type", "property_type", "neighbourhood"]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].str.lower().str.strip()
```

```python
df.head(2)
```

```python
if "price" in df.columns:
    q1 = df["price"].quantile(0.25)
    q3 = df["price"].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    df = df[(df["price"] >= lower) & (df["price"] <= upper)]
```

```python
df.head(2)
```

```python
# Price per guest
if "price" in df.columns and "accommodates" in df.columns:
    df["price_per_person"] = df["price"] / df["accommodates"]

# Host tenure (years)
if "host_since" in df.columns:
    df["host_tenure_years"] = 2026 - df["host_since"].dt.year

# Availability ratio
if "availability_365" in df.columns:
    df["availability_ratio"] = df["availability_365"] / 365
```

```python
df.head(1)
```

```python
drop_cols = [
    "thumbnail_url",
    "medium_url",
    "picture_url",
    "xl_picture_url"
]

df = df.drop(columns=[col for col in drop_cols if col in df.columns])
```
